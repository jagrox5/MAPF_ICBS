```python
import csv
import matplotlib.pyplot as plt

# Read the CSV file
csv_filename = "C:\\Users\\jagro\\Downloads\\MP ANuj\\MP ANuj\\instance_info.csv"

data = {'Nodes_Generated_Standard': [],
        'Nodes_Generated_Disjoint': [],
        'Nodes_Expanded_Standard': [],
        'Nodes_Expanded_Disjoint': [],
        'CPU_Time_Standard': [],
        'CPU_Time_Disjoint': []}

instance_files = []

with open(csv_filename, mode='r') as csv_file:
    reader = csv.DictReader(csv_file)
    for row in reader:
        instance_files.append(row['Instance_File'])
        data['Nodes_Generated_Standard'].append(int(row['Nodes_Generated_Standard']))
        data['Nodes_Generated_Disjoint'].append(int(row['Nodes_Generated_Disjoint']))
        data['Nodes_Expanded_Standard'].append(int(row['Nodes_Expanded_Standard']))
        data['Nodes_Expanded_Disjoint'].append(int(row['Nodes_Expanded_Disjoint']))
        data['CPU_Time_Standard'].append(float(row['CPU_Time_Standard']))
        data['CPU_Time_Disjoint'].append(float(row['CPU_Time_Disjoint']))
```

```python
# Plot 1: Nodes Generated comparison
plt.figure(figsize=(10, 4))
plt.plot(instance_files, data['Nodes_Generated_Standard'], label='Nodes Generated Standard', marker='o')
plt.plot(instance_files, data['Nodes_Generated_Disjoint'], label='Nodes Generated Disjoint', marker='x')
plt.xlabel('Instance Files')
plt.ylabel('Nodes Generated')
plt.title('Comparison of Nodes Generated (Standard vs Disjoint)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot 2: Nodes Expanded comparison
plt.figure(figsize=(10, 4))
plt.plot(instance_files, data['Nodes_Expanded_Standard'], label='Nodes Expanded Standard', marker='o')
plt.plot(instance_files, data['Nodes_Expanded_Disjoint'], label='Nodes Expanded Disjoint', marker='x')
plt.xlabel('Instance Files')
plt.ylabel('Nodes Expanded')
plt.title('Comparison of Nodes Expanded (Standard vs Disjoint)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot 3: CPU Time comparison
plt.figure(figsize=(10, 4))
plt.plot(instance_files, data['CPU_Time_Standard'], label='CPU Time Standard', marker='o')
plt.plot(instance_files, data['CPU_Time_Disjoint'], label='CPU Time Disjoint', marker='x')
plt.xlabel('Instance Files')
plt.ylabel('CPU Time')
plt.title('Comparison of CPU Time (Standard vs Disjoint)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
